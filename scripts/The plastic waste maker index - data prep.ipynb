```python
rank = [];producer =[];lst = [];dimensions=[]
rank = ([i for i in data["Rank"].values for j in range(5)])
producer = ([i for i in data["Polymer Producer"].values for j in range(5)])


for i,j,k,l,m in zip(data["No. of assets"], data["Production of in-scope polymers"], 
                     data["Flexible format contribution to SUP waste"], data["Rigid format contribution to SUP waste"],
                     data["Total contribution to SUP waste"]):
    lst.extend([i,j,k,l,m])
    

list1 = ['No. of assets', 'Production of in-scope polymers',
        'Flexible format contribution to SUP waste',
        'Rigid format contribution to SUP waste',
        'Total contribution to SUP waste'] 
dimensions = (list1*100)
```

```python
data_list = list(zip(rank, producer, dimensions, lst))
data_frame = pd.DataFrame(data_list, columns = ['Rank', 'Polymer Producer', 'dimension','value'])
```
